```python
import multiprocessing
import time
import psutil
import sys

def cpu_stress_task():
    """Function to run an infinite loop of computation."""
    while True:
        # A simple computation to keep the CPU busy
        _ = 1 * 1 

if __name__ == '__main__':
    processes = []
    # Determine the number of CPU cores to stress
    num_cores = psutil.cpu_count(logical=True)
    print(f"Starting {num_cores} processes to stress CPU...")

    for _ in range(num_cores):
        p = multiprocessing.Process(target=cpu_stress_task)
        p.start()
        processes.append(p)
    
    try:
        # Let the stress test run for a specified duration (e.g., 30 seconds)
        print("Stressing CPU for 30 seconds. Monitor your system's performance.")
        time.sleep(30) 
        print("Stress test complete.")

    except KeyboardInterrupt:
        print("Stress test interrupted by user.")
    finally:
        # Terminate all stress processes
        for p in processes:
            p.terminate()
        for p in processes:
            p.join()
        print("All processes terminated.")

```
